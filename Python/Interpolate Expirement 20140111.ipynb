```python
%pylab inline
import pandas as pd
#temps_file = './Katy/temps - Sheet1.csv'
temps_file = './Katy/nitro_test.csv'
original_data = pd.read_csv(temps_file, index_col=0, 
                            names=['date', 'temperature', None], 
                            usecols=['date', 'temperature'],
                            parse_dates=True)
print original_data.head()
print original_data.columns
original_data.plot()
print "Shape is:", original_data.shape
```

```python
data_15min = original_data.resample('15Min')
print data_15min.head()
```

```python
d15_interpol = data_15min.interpolate(method='time')
print d15_interpol.head()
d15_interpol.plot()
```

```python
out_file = './temps_15min_interpol.csv'
d15_interpol.to_csv(path_or_buf=out_file)
```
