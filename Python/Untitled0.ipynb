```python
import csv
from collections import OrderedDict
infile = './Data_for_Dan.csv'
theData = OrderedDict()
with open(infile, 'rb') as csvfile:
    csvreader = csv.reader(csvfile)
    for row in csvreader:
        if len(row) >= 2 and row[0] != 'Time':
            theData[row[0]] = []
            for val in row[1:]:
                try:
                    theData[row[0]].append(float(val))
                except:
                    theData[row[0]].append(None)
```

```python
badKeys = []
for key in theData.keys():
    if None in theData[key]:
        print key, theData[key]
        badKeys.append(key)
```
